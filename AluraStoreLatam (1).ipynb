### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total por tienda
ingresos_t1 = tienda["Precio"].sum()
ingresos_t2 = tienda2["Precio"].sum()
ingresos_t3 = tienda3["Precio"].sum()
ingresos_t4 = tienda4["Precio"].sum()
```

```python
print("Ingresos Tienda 1:", ingresos_t1)
```

```python
print("Ingresos Tienda 2:", ingresos_t2)
```

```python
print("Ingresos Tienda 3:", ingresos_t3)
```

```python
print("Ingresos Tienda 4:", ingresos_t4)
```

```python
import matplotlib.pyplot as plt

ingresos = [ingresos_t1, ingresos_t2, ingresos_t3, ingresos_t4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, ingresos, color="skyblue")
plt.title("Ingresos totales por tienda")
plt.ylabel("Ingresos (Suma de Precio)")
plt.show()
```

# 2. Ventas por categoría

```python
ventas_cat_t1 = tienda.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
ventas_cat_t2 = tienda2.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
ventas_cat_t3 = tienda3.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
ventas_cat_t4 = tienda4.groupby("Categoría del Producto")["Producto"].count().reset_index().sort_values(by="Producto", ascending=False)
```

```python
print("Ventas por categoría - Tienda 1")
print(ventas_cat_t1)
```

```python
print("Ventas por categoría - Tienda 2")
print(ventas_cat_t2)
```

```python
print("Ventas por categoría - Tienda 3")
print(ventas_cat_t3)
```

```python
print("Ventas por categoría - Tienda 4")
print(ventas_cat_t4)
```

# 3. Calificación promedio de la tienda


```python
# Calificaciones promedio de cada tienda
calif_t1 = tienda["Calificación"].mean()
calif_t2 = tienda2["Calificación"].mean()
calif_t3 = tienda3["Calificación"].mean()
calif_t4 = tienda4["Calificación"].mean()

print("Calificación promedio - Tienda 1:", calif_t1)
print("Calificación promedio - Tienda 2:", calif_t2)
print("Calificación promedio - Tienda 3:", calif_t3)
print("Calificación promedio - Tienda 4:", calif_t4)
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
prod_t1 = tienda.groupby("Producto")["Producto"].count().sort_values(ascending=False)
mas_vendidos_t1 = prod_t1.head(5)   # Top 5
menos_vendidos_t1 = prod_t1.tail(5) # Bottom 5

print("Más vendidos - Tienda 1:\n", mas_vendidos_t1)
print("\nMenos vendidos - Tienda 1:\n", menos_vendidos_t1)
```

```python
# Tienda 1 - Pie Chart
mas_vendidos_t1.plot(kind="pie", autopct='%1.1f%%', figsize=(6,6))
plt.title("Top 5 productos más vendidos - Tienda 1")
plt.ylabel("")  # quitamos el label de eje
plt.show()
```

```python
# Tienda 2
prod_t2 = tienda2.groupby("Producto")["Producto"].count().sort_values(ascending=False)
mas_vendidos_t2 = prod_t2.head(5)
menos_vendidos_t2 = prod_t2.tail(5)

print("\nMás vendidos - Tienda 2:\n", mas_vendidos_t2)
print("\nMenos vendidos - Tienda 2:\n", menos_vendidos_t2)
```
